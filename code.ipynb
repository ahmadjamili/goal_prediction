### lets go using label encoding

```python
from sklearn import preprocessing 
label_encoder2=preprocessing.LabelEncoder() 
label_encoder3=preprocessing.LabelEncoder() 
```

```python
dataset
```

```python
def body_part_encoding(x):
    num1=0
    num2=1
    num3=3
    if x=='پای راست' or x=='پای چپ':
        return num1
    if x=='سر':
        return num2
    if x=='سایر' :
        return num3
```

```python
dataset['bodyPart']=dataset['bodyPart'].apply(lambda x : body_part_encoding(x))
```

#### we have imbalanced data so these methods may help us

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()
x_train1, y_train1 = sm.fit_resample(x, y)
```

```python
from imblearn.over_sampling import RandomOverSampler
oversample = RandomOverSampler(sampling_strategy='minority')
x_train1, y_train1 = oversample.fit_resample(x, y)
```

```python
from imblearn.under_sampling import RandomUnderSampler
oversample = RandomUnderSampler(sampling_strategy='majority')
x_train1, y_train1 = oversample.fit_resample(x, y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_train1,y_train1,test_size=0.5)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 6)
classifier.fit(x_train, y_train)
prediction = classifier.predict(x_test)
```
